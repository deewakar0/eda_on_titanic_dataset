### Data Cleaning

```python
plt.figure(figsize=(12,7))
sns.set_style('whitegrid')
sns.boxplot('Pclass','Age',data=data)
```

### Here we use a basic python function which i just got by ctrl c + ctrl v (lol)  : see we could directly filled with average value but imputing average value in this manner gives the best method .;

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        else:
            return 24

    else:
        return Age
```

```python
data['Age'] = data[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(data.isnull(),yticklabels=False)
```

```python
data.drop('Cabin',axis=1,inplace=True)
```

```python
sns.heatmap(data.isnull(),yticklabels=False)
```

```python
sex = pd.get_dummies(data['Sex'],drop_first=True)
embark = pd.get_dummies(data['Embarked'],drop_first=True)
```

```python
data.drop(['Sex','Embarked','Name','Ticket','PassengerId'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data = pd.concat([data,sex,embark],axis=1)
```
